```python
sample = pd.read_csv('Data/train/303_P/303_COVAREP.csv', header=None)
```

```python
sample_1s = sample.groupby(sample['index']).mean()
```

```python
time = pd.read_csv('Data/train/303_P/303_TRANSCRIPT.csv', sep='\t')
```

```python
sample_1s['index'] = sample_1s.index
```

```python
time[time['speaker'] == 'Ellie']
```

```python
sample_1s.drop(sample_1s[sample_1s['index'].isin(range(int(time.loc[0,'start_time'])))].index, inplace = True)
for start,stop in zip(time['start_time'], time['stop_time']):
    sample_1s.drop(sample_1s[sample_1s['index'].isin(range(int(start),int(stop)))].index, inplace = True)
```

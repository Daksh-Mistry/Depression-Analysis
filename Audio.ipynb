## DataFrame

```python
train_labels = pd.read_csv('Data/train.csv')
test_labels = pd.read_csv('Data/dev.csv')
```

```python
df = []
for rank in train_labels['Participant_ID']:
    sample = pd.read_csv(f'Data/train/{rank}_P/{rank}_COVAREP.csv', header=None)
    sample['index'] = sample.index
    time = pd.read_csv(f'Data/train/{rank}_P/{rank}_TRANSCRIPT.csv', sep='\t')
    time = time[time['speaker'] == 'Ellie']
    for start,stop in zip(time['start_time'], time['stop_time']):
        sample.drop(sample[sample['index'].isin(range(int(100*start),int(100*stop)))].index, inplace = True)
    sample = sample[sample[1] == 1]
    df.append(sample.drop(columns = ['index']))
```

```python
df_test = []
for rank in test_labels['Participant_ID']:
    sample = pd.read_csv(f'Data/test/{rank}_P/{rank}_COVAREP.csv', header=None)
    sample['index'] = sample.index
    time = pd.read_csv(f'Data/test/{rank}_P/{rank}_TRANSCRIPT.csv', sep='\t')
    time = time[time['speaker'] == 'Ellie']
    for start,stop in zip(time['start_time'], time['stop_time']):
        sample.drop(sample[sample['index'].isin(range(int(100*start),int(100*stop)))].index, inplace = True)
    sample = sample[sample[1] == 1]
    df_test.append(sample.drop(columns = ['index']))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scale = pd.concat([sample.iloc[:,[0]] for sample in df])
sc.fit(scale)
```

```python
max_row = 0
```

```python
for sample in df:
    sample.iloc[:,[0]] = sc.transform(sample.iloc[:,[0]])
    sample.drop(columns=sample.columns[1], inplace = True)
    max_row = max(max_row, sample.shape[0])
```

```python
for sample in df_test:
    sample.iloc[:,[0]] = sc.transform(sample.iloc[:,[0]])
    sample.drop(columns=sample.columns[1], inplace = True)
    max_row = max(max_row, sample.shape[0])
```

```python
max_row
```

```python
df_reduce = []
df_test_reduce = []
for sample in df:
    df_reduce.append(sample.groupby(sample.index//25).mean())
for sample in df_test:
    df_test_reduce.append(sample.groupby(sample.index//25).mean())
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
X_train = pad_sequences(df_reduce, maxlen=3400, dtype='float32', padding='post', value=0.0)
X_test = pad_sequences(df_test_reduce, maxlen=3400, dtype='float32', padding='post', value=0.0)
```

```python
X_train.shape
```

```python
y_train = []
y_test = []
for label in train_labels['PHQ8_Binary']:
    y_train.append(label)
for label in test_labels['PHQ8_Binary']:
    y_test.append(label)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
y_train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Masking

model = Sequential([
    Masking(mask_value=0.0),
    LSTM(64, return_sequences=True),
    Dropout(0.3),
    LSTM(32),
    Dropout(0.3),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=4,
    validation_split=0.2
)
```

```python
pred = model.predict(X_test)
```

```python
pred
```

```python
y_pred = [1 if p > 0.5 else 0 for p in pred]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense, Dropout, Masking
from tensorflow.keras.callbacks import EarlyStopping

model2 = Sequential([
    Masking(mask_value=0.0),
    LSTM(32, return_sequences=True),
    Dropout(0.3),
    LSTM(16),
    Dropout(0.3),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])
model2.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
early_stop = EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)
```

```python
model2.fit(
    X_train, y_train,
    epochs=40,
    batch_size=4,
    validation_split=0.2,
    callbacks = [early_stop]
)
```

```python
pred = model2.predict(X_test)
```

```python
pred
```

```python
y_pred = [1 if p > 0.5 else 0 for p in pred]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
